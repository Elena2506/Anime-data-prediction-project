```python
# Created a copy of the dataframe
anime_1 = anime_rec.copy()
```

```python
# Replaced the NaN values on mediaType columm for 'Other'

anime_1['mediaType'].fillna('Other', inplace= True)
```

```python
anime_1.isna().sum()
```

```python
# I figure this code can be shorter. I have tried some many variations, but this one below was the only one that worked out. 
# How to know if this is the right solution?

anime_1['eps'].fillna(anime_1.groupby(['mediaType', 'studios'])['eps'].transform('median'),inplace= True)
anime_1['eps'].fillna(anime_1.groupby(['startYr', 'finishYr'])['eps'].transform('median'),inplace= True)
anime_1['eps'].fillna(anime_1.groupby(['ongoing', 'mediaType'])['eps'].transform('median'),inplace= True)

anime_1['duration'].fillna(anime_1.groupby(['mediaType', 'studios'])['duration'].transform('median'),inplace= True)
anime_1['duration'].fillna(anime_1.groupby(['startYr', 'finishYr'])['duration'].transform('median'),inplace= True)
anime_1['duration'].fillna(anime_1.groupby(['ongoing', 'mediaType'])['duration'].transform('median'),inplace= True)

anime_1['startYr'].fillna(anime_1.groupby(['mediaType', 'studios'])['startYr'].transform('median'),inplace= True)
anime_1['startYr'].fillna(anime_1.groupby(['startYr', 'finishYr'])['startYr'].transform('median'),inplace= True)
anime_1['startYr'].fillna(anime_1.groupby(['ongoing', 'mediaType'])['startYr'].transform('median'),inplace= True)

anime_1['finishYr'].fillna(anime_1.groupby(['mediaType', 'studios'])['finishYr'].transform('median'),inplace= True)
anime_1['finishYr'].fillna(anime_1.groupby(['startYr', 'finishYr'])['finishYr'].transform('median'),inplace= True)
anime_1['finishYr'].fillna(anime_1.groupby(['ongoing', 'mediaType'])['finishYr'].transform('median'),inplace= True)

anime_1['watched'].fillna(anime_1.groupby(['mediaType', 'studios'])['watched'].transform('median'),inplace= True)
anime_1['watched'].fillna(anime_1.groupby(['startYr', 'finishYr'])['watched'].transform('median'),inplace= True)
anime_1['watched'].fillna(anime_1.groupby(['ongoing', 'mediaType'])['watched'].transform('median'),inplace= True)

anime_1['rating'].fillna(anime_1.groupby(['mediaType', 'studios'])['rating'].transform('median'),inplace= True)
anime_1['rating'].fillna(anime_1.groupby(['startYr', 'finishYr'])['rating'].transform('median'),inplace= True)
anime_1['rating'].fillna(anime_1.groupby(['ongoing', 'mediaType'])['rating'].transform('median'),inplace= True)

anime_1['votes'].fillna(anime_1.groupby(['mediaType', 'studios'])['votes'].transform('median'),inplace= True)
anime_1['votes'].fillna(anime_1.groupby(['startYr', 'finishYr'])['votes'].transform('median'),inplace= True)
anime_1['votes'].fillna(anime_1.groupby(['ongoing', 'mediaType'])['votes'].transform('median'),inplace= True)
```
